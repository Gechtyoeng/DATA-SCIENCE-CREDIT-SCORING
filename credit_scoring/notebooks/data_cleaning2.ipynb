Handle Missing value

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("../data/raw/credit_risk_dataset.csv")
```

```python
df.isnull().sum()
```

Drop rows that have person_age > 65 (35 rows)  and <18 (0 rows)

```python
df = df.drop(df[(df['person_age'] < 18) | (df['person_age'] > 65)].index)
```

Drop rows that start working before 15 years old

```python
df = df.drop(
    df[(df['person_age'] - df['person_emp_length']) < 15].index
)
```

Drop loan_grade column

```python
df = df.drop('loan_grade', axis=1)
```

Drop row that has null in person_emp_length column

```python
df = df.drop(
    df[df['person_emp_length'].isna()].index
)
```

Fill missing values of loan_int_rate column with median

```python
df['loan_int_rate'] = df['loan_int_rate'].fillna(
    df['loan_int_rate'].median()
)
```

save data to a new csv file

```python
df.to_csv('../data/processed/cleaned_credit_risk_dataset.csv', index=False)
```
